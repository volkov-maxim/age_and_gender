```python
def datenum_to_year(datenum: np.uint32) -> np.uint16:
  """
  Convert Matlab datenum into Python datetime and return year.
  
  Args: 
    datenum (np.uint32): Date in datenum format (Matlab)
  
  Returns:
    np.uint16: Year corresponding to datenum
  """
  days = datenum % 1
  return (datetime.fromordinal(int(datenum)) \
         + timedelta(days=days) \
         - timedelta(days=366)).year
```

```python
ages = [photo_taken.item() - datenum_to_datetime(dob.item()) for (dob, photo_taken) in zip(mat_file["wiki"]["dob"][0][0][0], mat_file["wiki"]["photo_taken"][0][0][0])]
```
